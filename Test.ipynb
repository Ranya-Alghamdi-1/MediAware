```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Raniy/OneDrive/Desktop/genetic diseases/family_health_dataset.csv")
```

```python
df['bmi'] = df['Weight_kg'] / ((df['Height_cm'] / 100)**2)
```

```python
df.head()
```

```python
def classify_risk(row):
    bmi = row['bmi']
    output = []

    def risk_level(bmi):
        if bmi > 30:
            return 'High Risk'
        elif 25 <= bmi <= 30:
            return 'Moderate Risk'
        else:
            return 'Low Risk'

    if row['Has_Diabetes'] == 1:
        level = risk_level(bmi)
        if level != 'Low Risk':
            output.append(f"{level} of Diabetes")

    if row['Has_Hypertension'] == 1:
        level = risk_level(bmi)
        if level != 'Low Risk':
            output.append(f"{level} of Hypertension")

    if row['Has_Heart_Disease'] == 1:
        level = risk_level(bmi)
        if level != 'Low Risk':
            output.append(f"{level} of Heart Disease")

    return output if output else ["No concerning risks"]
```

```python
df['Risk_Assessment'] = df.apply(classify_risk, axis=1).apply(lambda x: ", ".join(x))
```

```python
def encode_risk_level(entry):
    if "High" in entry:
        return 2
    elif "Moderate" in entry:
        return 1
    else:
        return 0

df_dl['Risk_Label'] = df_dl['Risk_Assessment'].apply(encode_risk_level)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
# تحضير البيانات
# Encode Gender
df_dl['Gender_encoded'] = df_dl['Gender'].map({'Female': 0, 'Male': 1})
X = df_dl[['Age', 'bmi', 'Gender_encoded', 'Has_Diabetes', 'Has_Hypertension', 'Has_Heart_Disease']]
y = df_dl['Risk_Label']

# تحويل المخرجات إلى One-Hot Encoding
y_categorical = to_categorical(y, num_classes=3)

# تقسيم Train + Test
X_temp, X_test, y_temp, y_test = train_test_split(X, y_categorical, test_size=0.15, random_state=42)

# تقسيم Train + Validation
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```
